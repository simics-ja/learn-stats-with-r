# 仮説検定入門

## 仮説検定とは

__証明したいこと:__ 「提案手法Aは従来手法Bよりもxxで優れている」  

に対して，母集団の一部より得られたデータから母集団全体に対する仮説を正しいと主張するときに，研究の世界で広く受け入れられている統計的手法の一つ．  

<img src="./images/statistical-tests-image.png" width="640px">

## なぜ仮説検定が必要か

__データを集めた結果:__ 提案手法Aは従来手法Bよりもxxのデータがyy分だけ向上した!  

_しかし喜ぶのはまだ早い_

なぜなら，研究者は実験データに必ず誤差が生じることを知っている．  
yy向上したデータが___誤差なのか提案手法Aの影響なのか___を証明しない限り，その主張に納得する研究者はいない．

`「誤差であるか否か」に一定の結論を導くのが仮説検定を行う目的`

---

## 誤差には大きく2種類

誤差は測定値と真値の差であり，大きく2種類がある．

__偶然誤差:__ 同じ条件で何回実験を繰り返しても全く同じデータにはならないような誤差のこと．人の属性によらない個人差などが累積して生じたりする．  
__系統誤差:__ 実験器具や実験設定等で系統的に偏ってしまう誤差のこと．被験者属性の偏り（男だらけとか），実験実施者や被験者の先入観，実験に使ったデバイスなど．

仮説検定が評価するのは`偶然誤差`のほうであり，系統誤差はどうしようにもできない．  
系統誤差を減らすためには`実験計画法`のセオリーに沿って実験を行う必要がある．

## 誤差の確率分布

多くの場合に偶然誤差は正規分布（ガウス分布）に従うと仮定して評価する．
なぜなら，正規分布は平均値にその値を得られる確率のピークがあり，真値から離れた値ほど得られる確率が低いという特徴を持ち，偶然誤差の性質をうまく表現したモデルだから．

### 演習

Rで正規分布を書くプログラムを動かしましょう．
変数n,u,sの値を変えて正規分布の性質を確かめましょう．

* nを大きくするとグラフの形は？
 - 山なりの形が明確に
* nを大きくしていくと平均は？分散は？
 - 一定値に収束
* sを変えるとグラフの形は？
 - 横広がりに

```r
# n: 標本数（大きくしすぎると重くなるので注意）
n <- 10

# u: 平均値
u <- 50

# s: 標準偏差
s <- 10
data<-rnorm(n, mean=u, sd=s)
# hist()は引数からヒストグラム（x: 値， y: 出現頻度）を出力する関数
# rnormは与えた平均値と標準偏差の正規分布に従った疑似乱数をn個生成する関数
# 平均値uと標準偏差sで定義した正規分布の母集団から，n個の標本を集めることと同義
hist(data)
summary(data)
paste("分散は", sd(data))
```

このプログラムがやっていることは実験によってデータを得るという行為をシミュレーションしているのと同じ．  
私達が実験でやっていることは実験の被験者となる集団（ここでは正規分布と仮定）から，n人に被験者協力を依頼してデータの標本を集めるということ．  
ただし，現実で本当に正規分布になるかどうかは大量にデータを集めないとわからない．  
正規分布じゃなさそうなら，正規分布と仮定しない（ノンパラメトリック）検定を使う．

## 仮説検定の基本

「提案手法Aは従来手法Bよりも〇〇の指標で優れている」 と主張したい場合は大きく以下の流れになる．

参考: [仮説検定では、なぜまわりくどく、わかりにくい説明をするのか](https://toukeigaku-jouhou.info/2018/02/04/confusing-explanation-hypothesis-testing/)

### 1. 帰無仮説を立てる

対立仮説は帰無仮説が否定されたときに採択される仮説になる．

__帰無仮説:__ 「提案手法Aは従来手法Bと〇〇の差はない」  
* 否定するには一度でも誤りを見つければOK（簡単）

__対立仮説:__ 「提案手法Aは従来手法Bと〇〇の差がある」
* 肯定するには母集団全体を調べないといけない（困難）

### 2. 実験データを集める

優れていると主張する根拠である指標〇〇のデータを集める．  

### 3. 検定を行う

集めた各被験者データをまとめて検定にかける．  
検定ツールとしてRや [js-star](http://www.kisnet.or.jp/nappa/software/star/)などを使う．  

### 4. 帰無仮説を棄却する

検定ツールを使うとp値などの統計量が出力される．
p値が$ 0.0 < p <= 1.0$の値をとるので，この値が有意水準を下回っていれば帰無仮説を棄却できる．  
下回ったこと場合に手法間の平均に「有意な差がある」という．
有意水準は基本的に自分で決められるが，0.05に設定するのが一般的．  

### 5. 対立仮説に基づいて議論する

帰無仮説が正しいとは考えにくいので，対立仮説に基づいた考察を行う．　　

### 補足: 仮説検定の非対称性

仮説検定が結論づけるのは「帰無仮説が間違っている」という対立仮説に対する間接的な証明だけであり，対立仮説を直接証明しているわけではない．  
逆に，p値が有意水準を慕っ回っていなかったからといって，帰無仮説を正しいとすることもできない．
つまり，差がないことの証明に仮説検定は使えない．

__間違った結論の導き方の例:__ 有意な差はないので手法Aと手法Bは変わらない．

p値の意味については後述する．

### 演習

検定の基本中の基本である対応のあるt検定をやってみましょう．
method.Bをコメントアウトで切り替えて結果の違いを見てみましょう．

* 平均値の差がなくなるとt値やp値はどうなるか．
  - t値は0に近づき，p値は1に

```r
df<-data.frame(
    subject=c("A", "B", "C", "D", "E", "F", "G", "H", "I", "J"),
    method.A=c(0.7,-1.6,-0.2,-1.2,-0.1,3.4,3.7,0.8,0.0,2.0) ,
#      method.B=c(1.9, 0.8, 1.1, 0.1,-0.1,4.4,5.5,1.6,4.6,3.4) #有意差あり
#    method.B=c(1.2, 0.1, 1.0, 0.1,-0.1,4.4,2.2,0.1,1.1,0.4) # 有意差なし
  method.B=c(0.8,-1.7,-0.2,-1.2,-0.2,3.5,3.8,0.7,0.1,1.9)  # 有意差なし（method.Aと平均値は同じ）
)

print("検定にかけるデータは")
df
print("有意水準0.05で対応のあるt検定を行います．")

#t検定はt.test()
result<-t.test(x=df$method.A,y=df$method.B, paired=TRUE) # pairedは対応あり（被験者内），なし（被験者間）のフラグ
result
print(paste("手法Aの平均値", mean(df$method.A)))
print(paste("手法Bの平均値", mean(df$method.B)))
print(paste("p値は", result$p.value))
if (result$p.value < 0.05) {
    print("手法Aと手法Bに有意差があります．")
}
if (result$p.value >= 0.05) {
  print("手法Aと手法Bに有意差はありません．")
}
```

## p値の意味
### 有意差ありとはどういうことか
* 「帰無仮説が正しければ，5%以下の低い確率で生じる差を1回の実験で観測できた」
* 「100回実験やれば95回以上は出ないはずの差が1回の実験で観測できてしまった」

ゆえに帰無仮説が正しいという前提は間違いであると考える．

ちなみに，__「1回の実験で」__というところが重要ポイント．  
なぜなら，これらの前提だと100回実験すれば5回はまぐれで有意差ありになる可能性を許しているから．  
つまり，同じ実験を何度も繰り返せば有意差が出る可能性がその分だけ高くなる．  
後述の`検定の多重性`にもつながるお話．

### 有意差なしとはどういうことか
有意差ありの逆の考え方をするとわかりやすい．

* 「帰無仮説が正しければ，95%の確率で生じる差を観測できた」
* 「100回実験やれば95回観測できるであろう値を観測できた」

p値は帰無仮説が正しいと仮定したとき，どれだけ観測しにくい値であるかを意味し，帰無仮説の正しさを説明するものではない．  
つまり，__同じ実験を100回繰り返して実際に有意差ありになる回数が5回以下になるのか数えてみないことには証明できない．__  
なので，「有意差なし（p > 0.05）」は帰無仮説を全面的に肯定するものではない．

有意差がないので手法間に差はないと主張するのは結構恥ずかしい話なので注意．  
論文にどうしても書きたいのであれば「今回の実験では有意な差は確認できなかった」という前提をおいて，可能性について言及するレベルに留めておきましょう．

## 検定にかけられるデータの形

被験者にどの手法を割り当てるのか前もって考えてから実験しないと___検定できない形の表になる___ので要注意．  
t検定や後述のANOVAで検定できる形の表の例は以下の通り．

#### 被験者内計画の場合
被験者に対応がある（１人の被験者がA手法とB手法のどちらも実験を行う）場合の例は以下のようになる．  
検定を行うときに被験者個人の傾向を考慮する必要がある．

|---|---|---|

|kasanami|8.0|6.0|
|kurahashi|4.0|5.0|
|ishida|8.0|6.0|

#### 被験者間計画の場合
被験者に対応がない（１人の被験者は各手法のうちひとつだけ実験を行う）場合の例は以下のようになる．
被験者内よりも被験者集めがたいへん．

|method |s| value |
|---|---|---|
|A|yamada|10.0|
|A|kasanami|8.0|
|A|kurahashi|4.0|
|A|ishida|8.0|
|B|matsui|7.0|
|B|miki|3.0|
|B|luna|4.0|

#### 混合計画の場合
t検定の場合は基本的に1要因の2群間で検定を行うが，要因を2つ以上に増やして実験&検定を行う場合もある．  
例えば年齢を要因とするときは被験者内のようにどちらの要因にも振り分けるということができない．  
このように，被験者間計画の要因が含まれており，残りは被験者内計画の実験を混合計画という．  

| s | age | method:A | method:B |
|:-:|:-:|:-:|:-:|
| yamda | elder | 10.0 | 5.0 |

- __要因:__ method,age

### パラメトリック検定とノンパラメトリック検定

多くの仮説検定は偶然誤差がなんらかの分布（正規分布など）に従うという仮定の上で，実際に得られたデータがその仮定した誤差のモデル上でどれぐらいの確率で得られるデータなのか評価する．  このような仮説検定を__パラメトリック検定__という．  
逆に，分布を仮定せずに使える検定を__ノンパラメトリック検定__という．  
ノンパラメトリック検定のアイデアは平均値ではなく中央値などのデータの大小や順位だけで評価すること．  
外れ値に平均が引っ張られるようなデータに強い．  アンケートなどの非連続的で正規分布にはならないようなデータもノンパラメトリックを使うのが一般的．

## 検定の多重性

### 3群以上の検定

3群以上の検定で水準間の比較する場合，どのように検定するべきか．

|s| A | B | C |
|---|---|---|---|
|yamada|10.0|5.0|7.0|
|kasanami|8.0|6.0|6.0|
|kurahashi|4.0|5.0|5.0|
|ishida|8.0|6.0|2.0|

__ダメな方法:__ A-B,B-C,C-A間でt検定  
__一般的な方法:__ 分散分析&多重比較（ただし，分散分析を必要としない分散分析）

### 検定の多重性

何度もt検定を行うと，その分析で一度でも有意差を検出する可能性が上がってしまうことを`検定の多重性問題`という．  
上記の3群間の例では全ての群間で有意差なしになる確率が0.95^3=0.857に下がることになる．  
つまり，帰無仮説に基づいてるのに14%もの確率で有意差ありになることを許容しないといけなくなる．  

### 検定の多重性の解決方法

* 検定をどうにかひとつにする．
 - 一回の検定で仮説を証明できるようにデータ加工を行う．
* 検定に順序をつける．
 - 検定に順序をつけ，1度目の検定に有意差が出ないなら以降の検定を行わない．
   - いずれかが有意になるという状況を避けられる．
 - その順序に決めた妥当性については説明が必要になる．
* 有意水準を調整する．
 - Bonferroni法に代表されるような基準で有意水準を低く設定する．

参考: [多重性の調整方法は？統計学的検定の多重性を補正する３つの方法｜一番優しい、医薬品開発に必要な統計学の教本](https://best-biostatistics.com/multiple/avoid.html)

### 演習

検定の多重性をシミュレーションしましょう．

```r
n<-10
u<-50
s<-8

nOfTrial<-100

count=0

# 100回実験とt検定を繰り返す．
print(paste("実験とt検定を", nOfTrial, "回繰り返します．"))
for (i in 1:nOfTrial) {
    # 標本の取得
    # rnorm(n, mean=u, sd=s)で全く同じ分布からデータをn個取得
    df<-data.frame(
        A=rnorm(n, mean=u, sd=s),
        B=rnorm(n, mean=u, sd=s),
        C=rnorm(n, mean=u, sd=s)
    )

#
# 3群間のt検定で一度でも有意差が出た回数を数えるパターン
#  こっちなら14回前後有意差が検出される．
#    
    
    if (t.test(x=df$A, y=df$B, paired=TRUE)$p.value < 0.05 || t.test(x=df$B, y=df$C, paired=TRUE)$p.value < 0.05 || t.test(x=df$C, y=df$A, paired=TRUE)$p.value < 0.05) {
        count = count+1
    }

#
# 2群間のt検定で有意差を数えるパターン
#  こっちなら5回前後しか検出されない．
#

#     if (t.test(x=df$A, y=df$B, paired=TRUE)$p.value < 0.05) {
#         count = count+1
#     }
}
print("データの一例は以下になります．")
df
print(paste("一度でも有意差が検出された実験は", count, "回ありました．"))
```
